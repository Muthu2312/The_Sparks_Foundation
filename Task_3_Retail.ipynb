#Observation 4:

Profit Vs Discount

```python
fig,ax=plt.subplots(figsize=(20,8),sharex=True,sharey=True)
ax.scatter(df['Sales'],df['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df,palette="rainbow")
plt.legend()
plt.show()
```

There is no correlation between Discount and Profit

#Observaion 5:

Profit vs Quantity

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df, palette= "rocker_r")
plt.legend()
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['Black','Red'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

#Observation 6:

Profit and sales are maximum in consumer segment and minimum in Home Office segment

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='rainbow')
plt.xlabel('Region',fontsize=15)
plt.show()
```

#Observation 7:

Segment wise sales are almost same in every region

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['pink','Grey'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

#Observation 8:

Profit and sales are maximum in west region and minimum in south region

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['lightgreen','darkblue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

#Observation 9:



*   high profit is for california, new york
*   loss is for texas, pennsylvania, Ohio



```python
t_states = df['State'].value_counts().nlargest(10)
t_states
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['violet','purple'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

#Observation 10:

##As a business manager, try to find out the weak areas where you can work to make more profit?

*   Technology and Office Supplies have high profit
*   Furniture have less profit

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','lightblue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```
