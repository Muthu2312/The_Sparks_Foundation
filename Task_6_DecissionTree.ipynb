The purpose of this task is if we feed any new data to this classifier, it would be able to predict the right class accordingly

Required Libraries

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
import sklearn.metrics as sm
from sklearn.metrics import confusion_matrix
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import pydot
from IPython.display import Image
```

#Step 1: Import Dataset

```python
iris=pd.read_csv("/content/drive/MyDrive/The Sparks Foundation/Task 6_DecissionTree/Iris.csv")
```

#Step 3 - Data Visualization comparing various features

```python
sns.pairplot(iris)
```

#Step 4: Decision Tree Model Training

```python
iris.drop("Id",axis=1,inplace=True)
```

```python
X=iris.drop("Species",axis=1)
y=iris["Species"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1) 
tree_classifier = DecisionTreeClassifier()
tree_classifier.fit(X_train,y_train)
y_pred = tree_classifier.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

#Step 5: visualizing the Decision tree

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','versicolor','virginica']
export_graphviz(
        tree_classifier,
        out_file="img\desision_tree.dot",
        feature_names=fn,
        class_names=cn,
        rounded=True,
        filled=True
)

(graph,) = pydot.graph_from_dot_file('img\desision_tree.dot')
graph.write_png('img\desision_tree.png')

Image(filename='img\desision_tree.png') 
```

#Step 6 - Predicting the class output for some random values of petal and sepal length and width

```python
print("Class Names = ",cn)

# Estimating class probabilities
print()
print("Estimating Class Probabilities for flower whose petals length width are 4.7cm and 3.2cm and sepal length and width are 1.3cm and 0.2cm. ")
print()
print('Output = ',tree_classifier.predict([[4.7, 3.2, 1.3, 0.2]]))
print()
print("Our model predicts the class as 0, that is, Iris-setosa")
```

#Step 7 - Calculating the Model accuracy

```python
print("Accuracy:",sm.accuracy_score(y_test, y_pred))
```

The accuracy of this model is 1 or 100% since I have taken all the 4 features of the iris dataset for creating the decision tree model
